## Приоритизация гипотез

### Фреймворк ICE для приоритизации гипотез

```python
hypothesis['ICE'] = round((hypothesis['Impact'] * hypothesis['Confidence']) / hypothesis['Efforts'])
hypothesis[['Hypothesis', 'ICE']].sort_values(by='ICE', ascending = False)
```

Самые перспективные гипотезы:
 - запуск акции, дающей скидку на товар в день рождения (8 гипотеза),
 - добавление двух новых каналов привлечения трафика (0 гипотеза),
 - добавление формы подписки на все основные страницы для сбора базы клиентов (7 гипотеза)

### Фреймворк RICE для приоритизации гипотез

```python
hypothesis['RICE'] = round((hypothesis['Reach'] * hypothesis['Impact'] * hypothesis['Confidence']) / hypothesis['Efforts'])
hypothesis[['Hypothesis', 'RICE']].sort_values(by='RICE', ascending = False)
```

Самые перспективные гипотезы:
 - добавление формы подписки на все основные страницы для сбора базы клиентов (7 гипотеза),
 - добавление блоков рекомендаций товаров на сайт интернет магазина (2 гипотеза),
 - добавление двух новых каналов привлечения трафика (0 гипотеза)

### Изменение приоритизации гипотез при применении RICE вместо ICE

На первом месте фреймворка RICE гипотеза 7: добавление формы подписки на все основные страницы для сбора базы клиентов (в фреймворке ICE она была на 3 месте). Так произошло, т.к. параметр данной гипотезы Reach равен 10 (у других гипотез он меньше). Добавление формы подписки охватит большое количество пользователей интернет-магазина, поэтому такая гипотеза оценена максимальным баллом.

## Анализ A/B-теста

### График кумулятивной выручки по группам.

```python
# создадим таблицу groups с уникальными парами значений 'date' и 'group' таблицы orders. Избавимся от дубликатов
groups = orders[['date', 'group']].drop_duplicates()
```

```python
# агрегированные кумулятивные данные о заказах
orders_grouped = groups.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'],\
        orders['group'] == x['group'])].agg({'date': 'max',
                                            'group': 'max',
                                            'transactionId':pd.Series.nunique,
                                            'visitorId' : pd.Series.nunique,
                         'revenue': 'sum'}), axis = 1).sort_values(by =['date','group'])  
```

```python
# агрегированные кумулятивные данные о посетителях
visitors_grouped = groups.apply(lambda x: visitors[np.logical_and(visitors['date'] <= \
        x['date'], visitors['group'] == x['group'])].agg({'date': 'max',
                                                          'group': 'max',  
                                                          'visitors': 'sum'}), axis = 1).sort_values(by =['date','group'])
```

```python
# объединим сгруппированные таблицы
cumulativeData = orders_grouped.merge(visitors_grouped, left_on=['date', 'group'], right_on=['date', 'group'])
```

```python
# переименуем столбцы
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']
```

```python
# датафрейм с кумулятивным количеством заказов и кумулятивной выручкой по дням в группе А
cumulativeRevenueA  = cumulativeData[cumulativeData['group'] == 'A'][['date', 'revenue', 'orders']]
# датафрейм с кумулятивным количеством заказов и кумулятивной выручкой по дням в группе B
cumulativeRevenueB  = cumulativeData[cumulativeData['group'] == 'B'][['date', 'revenue', 'orders']]
```

```python
register_matplotlib_converters()
plt.figure(figsize=(10, 6))
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'], label = 'A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'], label = 'B')
plt.legend()
plt.ylim(0, 6000000)
plt.title('Кумулятивная выручка по группам')
plt.xlabel('Дата')
plt.ylabel('Выручка')
plt.xticks(rotation=70)
plt.show()
```

Выручка увеличивается в двух группах за исследуемый период. Выручка группы А растет равномерно. <br>
Выручка группы В имеет резкий скачок после 17.08.2019. Можно предположить, что в группе В есть аномалии и выбросы. <br>
По графику выручки можно сделать обширный вывод о причинах аномалий и выборосов. На резкое увеличение выручки группы В могут влиять, как аномально большие по стоимости (дорогие) заказы в выборке, так и большое количество небольших "обычных" заказов. 

### График кумулятивного среднего чека по группам.

```python
plt.figure(figsize=(10, 6))
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'] / cumulativeRevenueA['orders'], label='A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'] / cumulativeRevenueB['orders'], label='B')
plt.legend()
plt.title('Кумулятивный средний чек по группам')
plt.xlabel('Дата')
plt.ylabel('Средний чек')
plt.xticks(rotation=70)
plt.show()
```

Кумулятивное значение среднего чека группы А сначала колеблется, а после 13.08.2019 выравнивается, находится в диапазоне 7000-8000. <br>
Видим, что в группе В в диапазоне 17.08-21.08 есть аномалия - высокий средний чек. На высокий показатель среднего чека влияют аномально дорогие заказы. После выброса средний чек начинает снижаться.

### График относительного изменения кумулятивного среднего чека группы B к группе A.

```python
CumulativeRevenue_A_and_B = cumulativeRevenueA.merge(cumulativeRevenueB, left_on  = 'date',\
                                        right_on = 'date', how='left', suffixes=['A', 'B'])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(CumulativeRevenue_A_and_B['date'],
        (CumulativeRevenue_A_and_B['revenueB']/CumulativeRevenue_A_and_B['ordersB'])/
        (CumulativeRevenue_A_and_B['revenueA']/CumulativeRevenue_A_and_B['ordersA'])-1)
plt.axhline(y=0, color='grey', linestyle='--') 
plt.title('График относительного изменения кумулятивного среднего чека группы B к группе A')
plt.xlabel('Дата')
plt.ylabel('Относительное изменение')
plt.xticks(rotation=70)
plt.show()
```

В нескольких точках график различия между сегментами резко «скачет». Группа В увеличила средний чек относительно группы А. Есть 2 аномально дорогих заказов в группе В, из-за чего график сильно скачет. 

### График кумулятивной конверсии по группам.

```python
cumulativeData['conversion'] = cumulativeData['orders'] / cumulativeData['visitors']
```

```python
cumulativeData['conversion'] = cumulativeData['orders'] / cumulativeData['visitors']
cumulativeDataA = cumulativeData[cumulativeData['group'] == 'A']
cumulativeDataB = cumulativeData[cumulativeData['group'] == 'B']
```

```python
plt.figure(figsize=(10, 6))
plt.plot(cumulativeDataA['date'], cumulativeDataA['conversion'], label = 'A')
plt.plot(cumulativeDataB['date'], cumulativeDataB['conversion'], label = 'B')
plt.legend()
plt.title('Кумулятивная конверсия по группам')
plt.xlabel('Дата')
plt.ylabel('Конверсия')
plt.xticks(rotation=70)
plt.show()
```

Конверсия в группе А в диапазоне 01.08.2019 - 05.08.2019 была в районе 0.033 и выше. После 05.08.2019 наступил резкий спад, после чего конверсия продолжила снижаться. <br>
Конверсия в группе В колебалась и была выше конверсии группы А, начиная с 09.08.2019. <br>
Можно предположить, что в местах резкого роста конверсии, был резкое увеличение количества заказов. <br>
В начале теста конверсия по двум группам колебалась. После 09.08.2019 конверсия группы А стабилизировалась и находилась в диапазоне 0.029 - 0.031. Конверсия группы В в диапазоне: 0.033 - 0.036

### Относительное изменение кумулятивной конверсии группы B к группе A.

```python
CumulativeConversions = cumulativeDataA[['date','conversion']].merge(cumulativeDataB[['date','conversion']], \
                                    left_on='date', right_on='date', how='left', suffixes=['A', 'B'])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(CumulativeConversions['date'], CumulativeConversions['conversionB']/CumulativeConversions['conversionA']-1)
plt.axhline(y=0, color='black', linestyle='--')
plt.axhline(y=0.2, color='grey', linestyle='--')
plt.title('Относительное изменение кумулятивной конверсии группы B к группе A')
plt.xlabel('Дата')
plt.ylabel('Относительное изменение')
plt.xticks(rotation=70)
plt.show()
```

В начале теста конверсия группы В была выше конверсии группы А на 5%, затем наступил резкий спад, конверсия группы В снизилась на 8% по отношению к конверсии А. После 05.08.2019 конверсия группы В возрастает и продолжает расти до 17.08.2019. В диапазоне 13.08.2019 - 17.08.2019 разница между конверсиями А и В была около 20% (В лидирует). Позже наступил небольшой спад, группа В лидирует на 10-15% по конверсии относительно группы А.

### Статистическая значимость различий в конверсии между группами по «сырым» данным.

```python
orders_by_users_A = (orders[orders['group'] == 'A'].groupby('visitorId', \
                 as_index = False).agg({'transactionId': pd.Series.nunique}))
orders_by_users_A.columns = ['visitorId','orders']

orders_by_users_B = (orders[orders['group'] == 'B'].groupby('visitorId', \
                 as_index = False).agg({'transactionId': pd.Series.nunique}))
orders_by_users_B.columns = ['visitorId','orders']
```

```python
sampleA = pd.concat([orders_by_users_A['orders'], pd.Series(0, \
    index=np.arange(visitors[visitors['group']=='A']['visitors'].sum() - \
    len(orders_by_users_A['orders'])), name='orders')],axis=0)

sampleB = pd.concat([orders_by_users_B['orders'], pd.Series(0, \
    index=np.arange(visitors[visitors['group']=='B']['visitors'].sum() - \
    len(orders_by_users_B['orders'])), name='orders')],axis=0)
```

Гипотеза. Конверсии между группами А и В по «сырым» данным различаются. <br>
H0: Конверсия группы А равна конверсии группы В <br>
H1: Конверсия группы А не равна конверсии группы В

```python
alpha = .05 # критический уровень статистической значимости 
# это тот уровень, который мы принимаем для того, чтобы сказать, что мы считаем «значимым», а что нет
# если p-value окажется меньше него - отвергнем гипотезу

# применим критерий Манна-Уитни для сравнения двух независимых выборок 
# c параметром alternative = 'two-sided' для проверки двухсторонней гипотезы
results = stats.mannwhitneyu(sampleA, sampleB, alternative = 'two-sided')[1]

print('p-value: {0:.5f}'.format(results))

if (results < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")

print('Относительный прирост конверсии группы В к группе А по «сырым» данным: {0:.2%}'.format(
                                                        sampleB.mean() / sampleA.mean() - 1))
```

В ходе теста мы получили небольшое p-значение, что дает нам основание отвергнуть гипотезу Но о равенстве конверсий между группами. Можно сделать вывод, что конверсии между группами А и В по «сырым» данным имеют статистически значимые различия.

### Статистическая значимость различий в среднем чеке заказа между группами по «сырым» данным.

Гипотеза. Средний чек заказа между группами А и В по «сырым» данным различаются. <br>
H0: Средний чек заказа группы А равен среднему чеку группы В <br>
H1: Средний чек заказа группы А не равен среднему чеку группы В

```python
alpha = .05 # критический уровень статистической значимости 

results = stats.mannwhitneyu(orders[orders['group']=='A']['revenue'], orders[orders['group']=='B']['revenue'], \
                             alternative = 'two-sided')[1]

print('p-value: {0:.5f}'.format(results))

if (results < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")

print('Относительный прирост среднего чека группы B к группе А по «сырым» данным: {0:.2%}'.format(
    orders[orders['group']=='B']['revenue'].mean()/orders[orders['group']=='A']['revenue'].mean() - 1))
```

P-value больше 5%, статистически значимых отличий в среднем чеке между выборками нет.

### Статистическая значимость различий в конверсии между группами по «очищенным» данным.

```python
users_with_many_orders = pd.concat(
        [orders_by_users_A[orders_by_users_A['orders'] > 4]['visitorId'], 
        orders_by_users_B[orders_by_users_B['orders'] > 4]['visitorId']], axis = 0
)
```

```python
users_with_expensive_orders = orders[orders['revenue'] > 58233.2]['visitorId']
```

```python
ab_normal_users = (
        pd.concat([users_with_many_orders, users_with_expensive_orders], axis = 0)
        .drop_duplicates()
        .sort_values()
)
```

```python
sampleA_filtered = pd.concat(
    [orders_by_users_A[np.logical_not(orders_by_users_A['visitorId'].isin(ab_normal_users))]['orders'],\
     pd.Series(0, index=np.arange(visitors[visitors['group']=='A']['visitors'].sum() - \
     len(orders_by_users_A['orders'])), name='orders')],axis=0)

sampleB_filtered = pd.concat(
    [orders_by_users_B[np.logical_not(orders_by_users_B['visitorId'].isin(ab_normal_users))]['orders'],\
     pd.Series(0, index=np.arange(visitors[visitors['group']=='B']['visitors'].sum() - \
     len(orders_by_users_A['orders'])), name='orders')],axis=0)
```

Гипотеза. Конверсии между группами А и В по «очищенным» данным различаются. <br>
H0: Конверсия группы А равна конверсии группы В <br>
H1: Конверсия группы А не равна конверсии группы В

```python
alpha = .05 # критический уровень статистической значимости 

results = stats.mannwhitneyu(sampleA_filtered, sampleB_filtered, alternative = 'two-sided')[1]

print('p-value: {0:.5f}'.format(results))

if (results < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
    
print('Относительный прирост конверсии группы В к группе А по «очищенным» данным: {0:.2%}'.format(sampleB_filtered.mean() / \
                                                                                sampleA_filtered.mean() - 1))
```

В ходе теста мы получили небольшое p-значение, что дает нам основание отвергнуть гипотезу Но о равенстве конверсий между группами. Можно сделать вывод, что конверсии между группами А и В по «очищенным» данным имеют статистически значимые различия.

### Cтатистическая значимость различий в среднем чеке заказа между группами по «очищенным» данным.

Гипотеза. Средний чек заказа между группами А и В по «очищенным» данным различаются. <br>
H0: Средний чек заказа группы А равен среднему чеку группы В <br>
H1: Средний чек заказа группы А не равен среднему чеку группы В

```python
alpha = .05 # критический уровень статистической значимости 

results = stats.mannwhitneyu(orders[np.logical_and(orders['group'] == 'A', \
                                                   np.logical_not(orders['visitorId'].isin(ab_normal_users)))]['revenue'],
                            orders[np.logical_and(orders['group'] == 'B', \
                                                   np.logical_not(orders['visitorId'].isin(ab_normal_users)))]['revenue'], \
                                                   alternative = 'two-sided')[1]
print('p-value: {0:.5f}'.format(results))

if (results < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
    
print('Относительный прирост среднего чека группы B к группе А по «очищенным» данным: {0:.2%}'.format(
 orders[np.logical_and(orders['group'] == 'B', np.logical_not(orders['visitorId'].isin(ab_normal_users)))]['revenue'].mean() /
 orders[np.logical_and(orders['group'] == 'A', np.logical_not(orders['visitorId'].isin(ab_normal_users)))]['revenue'].mean()-1))
```

P-value больше 5%, статистически значимых отличий в среднем чеке между выборками нет. <br>
После исключения из данных аномально дорогих заказов, существует в среднем небольшая разница между группами по среднему чеку.
