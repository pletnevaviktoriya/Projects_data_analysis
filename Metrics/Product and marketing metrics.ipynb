### Продуктовые метрики

#### DAU, WAU и MAU
Вычислим средние значения этих метрик.

Выделим в отдельные столбцы дату, неделю, месяц и год по данным визитов на сайт.

```python
visits['session_date'] = visits['start_ts'].dt.date
visits['session_week'] = visits['start_ts'].dt.week
visits['session_month'] = visits['start_ts'].dt.month
visits['session_year'] = visits['start_ts'].dt.year
```

Сгруппируем данные по уникальным пользователям по дням/неделям/месяцу.

```python
# количество уникальных пользователей в день
dau = visits.groupby('session_date').agg({'uid': 'nunique'})
# количество уникальных пользователей в неделю
wau = visits.groupby(['session_year', 'session_week']).agg({'uid': 'nunique'})
# количество уникальных пользователей в месяц
mau = visits.groupby(['session_year', 'session_month']).agg({'uid': 'nunique'})
```

Посмотрим, как меняется количество уникальных пользователей по дням, неделям, месяцу.

```python
def func_graph (df, title, xname, yname):
    df.plot(figsize=(15,8), grid = True, legend = False)
    plt.title(title)
    plt.xlabel(xname)
    plt.ylabel(yname)
    plt.show()
func_graph(dau, 'Количество уникальных пользователей в день', 
          'День посещения', 'Количество пользователей')
```

```python
func_graph(wau, 'Количество уникальных пользователей в неделю', 
          'Неделя посещения', 'Количество пользователей')
```

```python
func_graph(mau, 'Количество уникальных пользователей в месяц', 
          'Месяц посещения', 'Количество пользователей')
```

```python
dau_total = dau.mean()
wau_total = wau.mean()
mau_total = mau.mean()
print('Среднее число пользователей в день (DAU):', int(dau_total))
print('Среднее число пользователей в неделю (WAU):', int(wau_total))
print('Среднее число пользователей в месяц (MAU):', int(mau_total))
```

По графикам видно, что в летний период (с июня по август 2017 года) количество пользователей не превышало 15000 в месяц. Предположим, что летом люди чаще уезжают отдыхать и реже посещают мероприятия. <br>
Начиная с сентября видим увеличение посещаемости сайта пользователями. <br>
Посещаемость сайта в новогодний период резко возросла. Можно предположить, что люди заранее планировали отдых и чаще покупали билеты в кино и театр в октябре 2017 года. <br>

Посмотрим на степень вовлеченности и лояльности пользователей. Вычислим sticky factor - показатель, который позволяет оценить регулярность посещений и стабильность пользовательской базы.

```python
sticky_mau = round((dau_total / mau_total) * 100, 2)
sticky_wau = round((dau_total / wau_total) * 100, 2)
print(sticky_wau)
print(sticky_mau)
```

Sticky factor показывает, с какой вероятностью новый привлеченный пользователь останется в продукте, насколько хорошо продукт удерживает и «цепляет» пользователей. <br>
Повторно к сервису обращается ~ 16% из 100 в месяц, 4% из 100 - в неделю.

#### Продолжительность типичной пользовательской сессии за весь период

```python
visits['duration_session'] = (visits['end_ts'] - visits['start_ts']).dt.seconds
```

```python
visits['duration_session'].hist(bins = 100, range = (0, 5000), figsize=(10,5))
plt.title('Продолжительность сессии')
plt.xlabel('Продолжительность')
plt.ylabel('Количество сессий')
plt.show()
```

Данное распределение продолжительности сессий похоже на распределение Пуассона. Найдем значение, которое чаще всего встречается в выборке (моду).

```python
mode = visits['duration_session'].mode()
print('Самая часто встречающаяся продолжительность сессии: {:.0f} секунд.'.format(mode[0]))
```

Чаще всего пользователь проводит на сайте 1 минуту. 

#### Retention Rate

Возьмём дату, когда пользователь впервые проявил активность на сайте.

```python
first_visit = visits.groupby('uid')['start_ts'].min()
first_visit.name = 'first_visit'
```

```python
# объединим таблицу first_visit (с датой первой активности на сайте) с исходым датафреймом
visits = visits.join(first_visit, on = 'uid')
```

```python
# выделим месяцы из столбцов first_visit и start_ts
visits['first_visit'] = visits['first_visit'].astype('datetime64[M]') # месяц первого посещения сайта (будущая когорта)
visits['visit_month'] = visits['start_ts'].astype('datetime64[M]')
```

```python
# рассчитаем lifetime пользователя в рамках когорты
visits['cohort_lifetime'] = visits['visit_month'] - visits['first_visit']
visits['cohort_lifetime'] = visits['cohort_lifetime'] / np.timedelta64(1, 'M')
visits['cohort_lifetime'] = visits['cohort_lifetime'].round().astype('int')
```

```python
# посчитаем для каждой когорты количество активных пользователей
cohorts = visits.groupby(['first_visit', 'cohort_lifetime']).agg({'uid': 'nunique'}).reset_index()
```

```python
# найдем исходное количество пользователей в когорте и переименуем столбец 
initial_cohorts_users = cohorts[cohorts['cohort_lifetime'] == 0][['first_visit','uid']]
initial_cohorts_users = initial_cohorts_users.rename(columns={'uid':'cohort_users'})
```

```python
# объединим данные по когортам с исходным количеством пользователей в когорте
cohorts = cohorts.merge(initial_cohorts_users,on='first_visit') 
```

```python
# рассчитаем Retention Rate
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users'] 
```

```python
# построем сводную таблицу, где для каждой когорты указана ее "продолжительность жизни" и рассчитан коэффициент удержания
retention_pivot = cohorts.pivot_table(index='first_visit',columns='cohort_lifetime',values='retention',aggfunc='sum') 
retention_pivot
```

```python
sns.set(style = 'white')
plt.figure(figsize = (15, 9))
plt.title('Retention Rate')
sns.heatmap(retention_pivot, annot=True, fmt='.2%', linewidths=1, linecolor='gray', vmin=0.01, vmax=0.1,
            yticklabels = retention_pivot.index.date)
plt.xlabel('Период жизни когорты')
plt.ylabel('Когорта')
plt.show()
```

```python
print('Cредний Retention Rate на второй месяц «жизни» когорт: {:.2%}'.format(retention_pivot[2].mean()))
```

Коэффициент удержания довольно высок у "летних" когорт (июнь/июль/август) на следующий после покупки месяц пользования сайтом. Можно предположить, что это обусловлено тем, что начинается летний период, и пользователи чаще уходят в отпуск, у них остается больше времени на посещение мероприятий. <br>
У осенних когорт коэффициент удержания также высок (на уровне летних и даже выше - у когорты в сентябре retention более 8%). Предположим, что пользователи начинают планировать новогодние праздники и заранее покупают билеты на мероприятия. <br>
Январская и весенние когорты показывают ниже коэффициенты, чем предыдущие когорты. <br>
На первый месяц жизни всех когорт остается примерно от 4 до 8.5% пользователей (более 90% пользователей сайт теряет уже на следующий месяц пользования ресурсом). Данный факт можно объяснить спецификой продукта. Так как сайт в основном подразумевает продажу билетов на мероприятия (кино, театр, концерты), не каждый пользователь будет покупать билет каждый месяц. <br>

Можно предположить, что это обусловлено тем, что начинается летний период, и пользователи чаще уходят в отпуск, у них остается больше времени на посещение мероприятий.
У осенних когорт коэффициент удержания также высок (на уровне летних и даже выше - у когорты в сентябре retention более 8%). Предположим, что пользователи начинают планировать новогодние праздники и заранее покупают билеты на мероприятия.
Январская и весенние когорты показывают ниже коэффициенты, чем предыдущие когорты.
На первый месяц жизни всех когорт остается примерно от 4 до 8.5% пользователей (более 90% пользователей сайт теряет уже на следующий месяц пользования ресурсом).

#### Среднее количество покупок на одного покупателя за 6 месяцев

```python
# объединим таблицу заказов с таблицей с датой, когда пользователь впервые приобрел билет на сайте 
orders = orders.merge(first_order_date, on = 'uid')
```

```python
# выделим месяцы из столбцов first_order и buy_ts
orders['first_order_month'] = orders['first_order'].astype('datetime64[M]')
orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')
```

```python
# с помощью группировки создадим таблицу когорт 
orders_cohorts = (orders.groupby(['first_order_month','order_month'])
                        .agg({'uid':['nunique','count'], 'revenue': 'sum'})
                        .reset_index())
orders_cohorts.columns = ['first_order_month','order_month','users_cnt','orders_cnt','revenue_sum']
```

```python
orders_cohorts['cohort_lifetime'] = orders_cohorts['order_month'] - orders_cohorts['first_order_month']
orders_cohorts['cohort_lifetime'] = orders_cohorts['cohort_lifetime'] / np.timedelta64(1,'M')
orders_cohorts['cohort_lifetime'] = orders_cohorts['cohort_lifetime'].round().astype('int')
```

```python
# найдем исходное количество пользователей в когорте и переименуем столбец 
orders_initial_cohorts = orders_cohorts[orders_cohorts['cohort_lifetime'] == 0][['first_order_month','users_cnt']]
orders_initial_cohorts = orders_initial_cohorts.rename(columns={'users_cnt':'cohort_users'})
```

```python
orders_cohorts = orders_cohorts.merge(orders_initial_cohorts,on='first_order_month') 
```

```python
# найдем количество покупок на каждого пользователя
orders_cohorts['order_per_user'] = orders_cohorts['orders_cnt']/orders_cohorts['users_cnt']
```

```python
order_pivot_table = orders_cohorts.pivot_table(index='first_order_month',columns='cohort_lifetime',
                                      values='order_per_user',aggfunc='mean') 
order_pivot_table.loc['mean'] = order_pivot_table.mean()
order_pivot_table
```

```python
y_name = ['2017-06-01', '2017-07-01', '2017-08-01', '2017-09-01', '2017-10-01', '2017-11-01', '2017-12-01',
    '2018-01-01', '2018-02-01', '2018-03-01', '2018-04-01', '2018-05-01', '2017-06-01', 'mean']
```

```python
sns.set(style = 'white')
plt.figure(figsize = (15, 9))
plt.title('Количество покупок на одного покупателя')
sns.heatmap(order_pivot_table, annot=True, fmt='.2f', linewidths=1, linecolor='gray', 
            yticklabels = y_name)
plt.xlabel('Период жизни когорты')
plt.ylabel('Когорта')
plt.show()
```

```python
print('Среднее количество покупок на одного покупателя за 6 месяцев: {:.2f} '.\
      format(order_pivot_table.cumsum(axis = 1).mean(axis=0)[5]))
```

В среднем 1 пользователь совершает 1-2 покупки билетов в месяц, что выглядит достаточно правдоподобно. Получается, за полгода пользователь совершает в среднем 10-11 покупок.

#### LTV на покупателя по когортам

```python
# маржинальность сервиса - 100%
margin_rate = 1
orders_cohorts['gp'] = orders_cohorts['revenue_sum'] * margin_rate
# разделим валовую прибыль когорт за каждый месяц на общее число пользователей в каждой когорте
orders_cohorts['ltv'] = orders_cohorts['gp'] / orders_cohorts['cohort_users']
```

```python
ltv_pivot_table = orders_cohorts.pivot_table(index = 'first_order_month', columns = 'cohort_lifetime',
                                       values = 'ltv', aggfunc = 'mean')
```

```python
# LTV — накопительная метрика
ltv_pivot_table = ltv_pivot_table.cumsum(axis=1)
```

#### ROMI по когортам в разрезе источников

Чтобы посчитать ROMI по рекламных источникам, необходимо найти LTV в разрезе источников.

```python
# найдем количество покупателей в разрезе рекламного источника
cohort_sizes = visits_orders.groupby(['first_order_month', 'source_id']).agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'source_id', 'n_buyers']
cohort_sizes.head()
```

```python
# найдем выручку по каждому рекламному источнику по когортам
source_cohorts = visits_orders.groupby(['first_order_month','order_month', 'source_id']).agg({'revenue': 'sum'}).reset_index()
source_cohorts.head()
```

```python
# объединим 2 таблицы
result_source = cohort_sizes.merge(source_cohorts, on=['first_order_month', 'source_id'])
```

```python
result_source.head()
```

```python
result_source['lifetime'] = result_source['order_month'] - result_source['first_order_month']
result_source['lifetime'] = result_source['lifetime'] / np.timedelta64(1, 'M')
result_source['lifetime'] = result_source['lifetime'].round().astype('int')
```

```python
result_source['ltv'] = result_source['revenue'] / result_source['n_buyers']
```

```python
result_source.head()
```

```python
# средний САС
cac_mean = cac_source_month.groupby('source_id')['cac'].mean()
cac_mean
```

```python
romi = result_source.merge(cac_mean, on = 'source_id')
```

```python
# найдем ROMI
romi['romi'] = romi['ltv'] / romi['cac']
romi.head()
```

```python
# все типы рекламных источников
sources = set(romi['source_id']) 
sources
```

```python
# напишем функцию для расчета ROMI по каждому рекламному источнику
def romi_heatmap(source):
    romi_cohort = romi[romi['source_id'] == source]
    romi_cohort_pivot = romi_cohort.pivot_table(index='first_order_month', columns='lifetime', 
        values='romi', aggfunc='mean')
    romi_cohort_pivot = romi_cohort_pivot.cumsum(axis=1).round(2) 
    y_name = ['2017-06-01', '2017-07-01', '2017-08-01', '2017-09-01', '2017-10-01', '2017-11-01', '2017-12-01',
    '2018-01-01', '2018-02-01', '2018-03-01', '2018-04-01', '2018-05-01', '2017-06-01']
    sns.set(style = 'white')
    plt.figure(figsize = (15, 9))
    plt.title('ROMI по источнику: {}'.format(source))
    sns.heatmap(romi_cohort_pivot, annot=True, linewidths=1, linecolor='gray', yticklabels = y_name)
    plt.xlabel('Период жизни когорты')
    plt.ylabel('Когорта')
    plt.show()
```

```python
for source in sources:
    romi_heatmap(source)
```

**ROMI по 1 источнику.** Летние когорты окупились уже на 3 месяц (ROMI более 1). Осенние когорты окупились на 6-7 месяц или близки к этому. <br>
**ROMI по 2 источнику.** Когорты сентября и декабря окупились на 2-3 месяцы. Когорты июня окупилась на 11 месяц. Остальные когорты не окупились, показывают слабую динамику прироста. Можно предположить, что люди приходили с этого рекламного источника и делали покупки в связи с сезонностью. Общая картина такова, что почти все когорты не окупаются. <br>
**ROMI по 3 источнику.** Ни одна когорта не окупилась! Помним, что в этот источник вложено больше всего маркетинговых средств. Но пользователи мало покупают. <br>
