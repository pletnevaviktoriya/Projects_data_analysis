# Анализ оттока клиентов фитнес-клуба 
Сеть фитнес-центров «Культурист-датасаентист» разрабатывает стратегию взаимодействия с клиентами на основе аналитических данных. Для фитнес-центра можно считать, что клиент попал в отток, если за последний месяц ни разу не посетил спортзал. Чтобы бороться с оттоком, отдел по работе с клиентами «Культуриста-датасаентиста» перевёл в электронный вид множество клиентских анкет. Необходимо провести анализ и подготовить план действий по удержанию клиентов.

## Загрузка данных, изучение общей информации

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score
from scipy.cluster.hierarchy import dendrogram, linkage 
from sklearn.cluster import KMeans
```

```python
gym_churn = pd.read_csv('/datasets/gym_churn.csv')
```

```python
# сделаем строчные заголовки столбцов
gym_churn.columns = gym_churn.columns.str.lower()
```

```python
gym_churn.info()
```

Посмотрим на описание и типы данных во всех колонках.
- `churn` — факт оттока в текущем месяце, целые числа; <br>

Данные клиента за предыдущий до проверки факта оттока месяц:
 - `gender` — пол, целые числа;
 - `near_location` — проживание или работа в районе, где находится фитнес-центр, целые числа;
 - `partner` — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент — в таком случае фитнес-центр хранит информацию о работодателе клиента), целые числа;
 - `promo_friends` — факт первоначальной записи в рамках акции «приведи друга» (использовал промо-код от знакомого при оплате первого абонемента), целые числа;
 - `phone` — наличие контактного телефона, целые числа;
 - `age` — возраст, целые числа;
 - `lifetime` — время с момента первого обращения в фитнес-центр (в месяцах), целые числа. <br>
 
Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:
 - `contract_period` — длительность текущего действующего абонемента (месяц, 3 месяца, 6 месяцев, год), целые числа;
 - `month_to_end_contract` — срок до окончания текущего действующего абонемента (в месяцах), вещественные числа;
 - `group_visits` — факт посещения групповых занятий, целые числа;
 - `avg_class_frequency_total` — средняя частота посещений в неделю за все время с начала действия абонемента, вещественные числа;
 - `avg_class_frequency_current_month` — средняя частота посещений в неделю за предыдущий месяц, вещественные числа;
 - `avg_additional_charges_total` — суммарная выручка от других услуг фитнес-центра: кафе, спорт-товары, косметический и массажный салон, вещественные числа.

```python
# с целью экономии памяти заменим некоторые столбцы на целые числа размером 1 байт
for column in ['gender', 'near_location', 'partner', 'promo_friends', 'phone', \
               'contract_period', 'group_visits', 'age', 'lifetime', 'churn']:
    gym_churn[column] = gym_churn[column].astype('int8')
```

## Модель прогнозирования оттока клиентов

```python
# разделим данные на признаки (матрица X) и целевую переменную (y)
X = gym_churn.drop('churn', axis = 1)
y = gym_churn['churn']
```

```python
# разделим модель на обучающую и валидационную выборку
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# стандартизируем признаки
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train)
X_test_st = scaler.transform(X_test)
```

### Логистическая регрессия

```python
# зададим алгоритм для модели логистической регрессии
lr_model = LogisticRegression(random_state=0, solver='liblinear')
# обучим модель
lr_model.fit(X_train_st, y_train)
# воспользуемся уже обученной моделью, чтобы сделать прогнозы
lr_predictions = lr_model.predict(X_test_st)
lr_probabilities = lr_model.predict_proba(X_test_st)[:, 1]
```

```python
def print_all_metrics(y_true, y_pred, y_proba, title='Метрики классификации'):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
```

```python
print_all_metrics(
    y_test,
    lr_predictions,
    lr_probabilities,
    title='Метрики для модели логистической регрессии:',
)
```

### Случайный лес

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=0)
# обучим модель случайного леса
rf_model.fit(X_train_st, y_train)
# воспользуемся обученной моделью, чтобы сделать прогнозы
rf_predictions = rf_model.predict(X_test_st)
rf_probabilities = rf_model.predict_proba(X_test_st)[:, 1]
# выведем все метрики
print_all_metrics(
    y_test,
    rf_predictions,
    rf_probabilities,
    title='Метрики для модели случайного леса:',
)
```

Алгоритм логистической регрессии показывает лучше результаты по метрикам, чем случайный лес. Доля правильных ответов для модели логистической регрессии равна 90%, точность равна 79%, полнота - 81%.

**Вывод**
1. Разбили данные на обучающую и валидационную выборку.
2. Стандартизировали признаки.
3. Обучили модель на train-выборке двумя способами: логистической регрессией и случайным лесом.
4. Оценили метрики accuracy, precision и recall для обеих моделе.
5. Алгоритм логистической регрессии показывает лучше результаты по метрикам, чем случайный лес. Доля правильных ответов для модели логистической регрессии равна 90%, точность равна 79%, полнота - 81%.

## Кластеризация клиентов

### Дендрограмма

```python
X_st = scaler.fit_transform(X)
```

```python
linked = linkage(X_st, method = 'ward') 
```

```python
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('Дендрограмма')
plt.show() 
```

На основании графика можно предположить, что всего имеем 5 кластеров.

### Кластеризация на основании алгоритма K-Means

```python
# задаём модель k_means с числом кластеров 5
km = KMeans(n_clusters = 5, random_state=0)
# прогнозируем кластеры для наблюдений (алгоритм присваивает им номера от 0 до 2)
labels = km.fit_predict(X_st)
# сохраняем метки кластера в поле нашего датасета
gym_churn['cluster'] = labels
```

### Средние значения признаков для кластеров

```python
# выводим статистику по средним значениям наших признаков по кластеру
claster = gym_churn.groupby('cluster').mean().T
claster
```

1. Наибольший отток (почти 40%) составляют посетители 0 кластера. Можно заметить, что  для таких клиентов не свойственно проживание или работа в районе, где находится фитнес-центр.
2. Далее следует 2 кластер по оттоку посетителей. Для таких посетителей не характерно посещение групповых программ.
3. Третье место по оттоку занимает 4 кластер. Отсутсвие контакного номера телефона - характерный признак оттока такого кластера.
